```python
puffer_oben = df['puffer_oben'].to_numpy()
puffer_unten = df['puffer_unten'].to_numpy()
kollektor = df['col'].to_numpy()
t = df['t'].to_numpy()
```

```python
%matplotlib inline 
from matplotlib import pyplot as pl
pl.rcParams["figure.figsize"] = (10,5)
x = range(len(puffer_oben))
```

```python
pl.plot(t,puffer_oben,label="po")
pl.plot(t,puffer_unten,label="pu")
pl.plot(t,kollektor,label="col")
pl.xticks(np.arange(min(x), max(x)+1, 100))
pl.gcf().autofmt_xdate()
pl.grid()
pl.legend()
pl.show()
```
